```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('platosim_SPF.dat', delim_whitespace=True) # Setting delimeter as whitespace
df_top = df.head()
#print(df.iloc[0])
```

```python
reqd_Index = df[df['logTe']<=3.5].index.tolist()
#print(reqd_Index) #row numbers for stars with temp < 3500 K
```

```python
indices = []   #index values for stars with Temp < 3500 K and mass between 0.08 to 0.6 solar masses 

for i in reqd_Index:
    mass = (df["Mass"].iloc[i])
    if (0.08 < mass < 0.6):
        indices.append(i)
        
#print(indices)
```

```python
df_2 = (df.iloc[indices])

ax1 = df_2.plot.scatter(x='logTe', y='logL', c='darkblue')
plt.xlim(plt.xlim()[::-1])

plt.xlabel('log T_eff')
plt.ylabel('log L')

plt.title('HR diagram for stars with T<3500K and between 0.08 and 0.6 solar masses ')
```
